```python
import pandas as pd
import numpy as np
```

```python
# Считываем исходные данные
# Для удобства проверки задания и возможности прогона скрипта зальём датасеты на Google.Drive.
url = 'https://docs.google.com/spreadsheets/d/1H-FIBAiCiBCFVlBwCBo4WGQjXsUEnQd_/edit?usp=share_link&ouid=104521593033447016990&rtpof=true&sd=true'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
initial_data = pd.read_excel(path, sheet_name=0)
```

### Решение

```python
# 1. Убираем записи, состоящие целиком из NAN значений, т.к. маловероятно, что они нам пригодятся.
df = initial_data.dropna(how='all')
```

```python
# 2. Имеются столбцы, состоящие целиком из NAN значений. Удалим их, чтобы не мозолили глаз.
df = df.dropna(axis=1, how='all')
```

```python
df
```

```python
# 3. Для удобства переименуем колонки
df = df.rename(columns={'№': 'fio', 
                   'Диагноз': 'diagnosis', 
                   'Код \nуслуги': 'service_code', 
                   'Название услуги': 'tittle', 
                   'Цена по прейскуранту': 'price', 
                   'Цена со скидкой': 'discounted_price',
                   'Кол-\nво': 'quantity',
                   'Сумма со скидкой': 'discounted_sum'})  
```

```python
# 4. Создадим колонку 'Номер полиса' ('policy_number')
df['policy_number'] = df.fio
```

```python
df
```

```python
# 5. Займёмся атрибутом ФИО (fio)
# Создадим функцию split_fio, которая отбирает элементы [0:3] от строки (fio) и объединяет их.
def split_fio(fio):
    return ' '.join(str(fio).split(' ')[0:3])
```

```python
df['fio'] = df.fio.apply(split_fio)
```

```python
# 6. По аналогии поступим и созданным полем 'policy_number'
def split_policy_number (policy_number):
    return str(policy_number).split(' ')[-1]
```

```python
df['policy_number'] = df.policy_number.apply(split_policy_number)
```

```python
df
```

```python
# 7. Создадим пустую колонку 'Дата оказания услуги' (date). 
# Данный шаг совсем не обязателен и может быть пропущен, однако с ним визуально проще сравнивать нашу таблицу с ожидаемым результатом. 
df['date'] = np.nan
```

```python
df.columns
```

```python
# Расположим столбцы в требуемом порядке и обновим индексы.
order_of_columns = ['fio', 'policy_number', 'date', 'service_code', 'diagnosis', 'tittle', 'quantity', 'price', 'discounted_sum']
df = df[order_of_columns]
df = df.reset_index(drop=True)
df
```

```python
# Посмотрим на первое значение столбца fio.
df.fio.iloc[0]
```

```python
# 8. Составим цикл для замены 'nan' на полные ФИО. 
# Плюс такого подхода в том, что он будет работать при изменении количества записей и их последовательности. 
# Минус - скорость работы на больших данных.
names = []
x = 0
n = 0
for i in df.fio: 
    if i != 'nan' and i == df.fio.iloc[x]:
        n += 1
        names.append(i)
    elif i == 'nan':
        i = df.fio.iloc[x]
        n += 1
        names.append(i)
    elif i != 'nan':
        x = n
        i = df.fio.iloc[x]
        n += 1
        names.append(i)

df['fio'] = names
```

```python
# 9. Аналогично поступим с policy_number

pls_numbers = []
x = 0
n = 0
for i in df.policy_number: 
    if i != 'nan' and i == df.policy_number.iloc[x]:
        n += 1
        pls_numbers.append(i)
    elif i == 'nan':
        i = df.policy_number.iloc[x]
        n += 1
        pls_numbers.append(i)
    elif i != 'nan':
        x = n
        i = df.policy_number.iloc[x]
        n += 1
        pls_numbers.append(i)

df['policy_number'] = pls_numbers
```

```python
df
```

```python
# 10. Из вышеприведённой таблицы видно, что местоположение даты имеет закономерность. Она выше названия услуги на 1 строку.
# Заполним колонку 'date' с соответствующим смещением.
df['date'] = df.diagnosis.shift(1)
#df
```

```python
# 11. Удалим все строки, имеющие хотя бы одно NAN.
df = df.dropna()
df
```

```python
# 12. Обновим индекс
df = df.reset_index(drop=True)
```

```python
# 13. Приведём 3 последних столбца к требуемому виду.
df.price = df.price.astype(int)
df.discounted_sum = df.discounted_sum.astype(int)
df.quantity = df.quantity.astype(int)
```

```python
df
```

```python
# 14. Финальное переименование
df = df.rename(columns={'fio': 'ФИО', 
                   'policy_number': 'Номер полиса', 
                   'date': 'Дата оказания услуг', 
                   'service_code': 'Код услуги', 
                   'diagnosis': 'Диагноз', 
                   'tittle': 'Наименование услуги',
                   'quantity': 'Кол-во',
                   'price': 'Цена по прейскуранту', 
                   'discounted_sum': 'Сумма со скидкой'})  
```

```python
# Результат
df
```
