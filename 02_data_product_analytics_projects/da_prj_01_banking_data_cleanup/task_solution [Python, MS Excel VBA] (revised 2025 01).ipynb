С помощью скрипта на python преобразовать исходные данные до датафрейма требуемого формата (ожидаемый результат представлен на листе «Вывод»)

```python
import pandas as pd
import numpy as np
```

```python
# Считываем исходные данные
# Для удобства проверки задания и возможности прогона скрипта зальём датасеты на Google.Drive.
url = 'https://docs.google.com/spreadsheets/d/1H-FIBAiCiBCFVlBwCBo4WGQjXsUEnQd_/edit?usp=share_link&ouid=104521593033447016990&rtpof=true&sd=true'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
initial_data = pd.read_excel(path, sheet_name=0)

initial_data
```

```python
# Посмотрим типы данных
initial_data.dtypes
```

```python
# Считываем данные и отображаем ожидаемый результат
target_df = pd.read_excel(path, sheet_name=1)
target_df
```

### Решение

```python
# 1. Убираем записи и колонки, состоящие целиком из NAN значений, т.к. маловероятно, что они нам пригодятся.
df = initial_data.dropna(how='all').dropna(axis=1, how='all')
```

```python
# 3. Для удобства переименуем колонки
df = df.rename(columns={'№': 'fio', 
                   'Диагноз': 'diagnosis', 
                   'Код \nуслуги': 'service_code', 
                   'Название услуги': 'tittle', 
                   'Цена по прейскуранту': 'price', 
                   'Цена со скидкой': 'discounted_price',
                   'Кол-\nво': 'quantity',
                   'Сумма со скидкой': 'discounted_sum'})  
```

```python
# 4. Создадим колонку 'Номер полиса' ('policy_number') на основании столбца fio
df['policy_number'] = df.fio

# df.head()
```

```python
# 5. Займёмся атрибутом ФИО (fio)
# Создадим функцию split_fio, которая отбирает элементы [0:3] от строки (fio) и объединяет их.

df['fio'] = df.fio.apply(lambda x: x if pd.isna(x) else ' '.join(str(x).split(' ')[0:3]))
```

```python
# 6. По аналогии поступим и созданным полем 'policy_number'
df['policy_number'] = df.policy_number.apply(lambda x: x if pd.isna(x) else str(x).split()[-1])
```

```python
df
```

```python
# 7. Создадим пустую колонку 'Дата оказания услуги' (date). 
# Данный шаг совсем не обязателен и может быть пропущен, однако с ним визуально проще сравнивать нашу таблицу с ожидаемым результатом. 
df['date'] = None
```

```python
# Расположим столбцы в требуемом порядке и обновим индексы.
order_of_columns = ['fio', 'policy_number', 'date', 'service_code', 'diagnosis', 'tittle', 'quantity', 'price', 'discounted_sum']
df = df[order_of_columns].reset_index(drop=True)
df
```

```python
# Заполним пропуски последним известным значением сверху

df['fio'] = df['fio'].ffill()
df['policy_number'] = df['policy_number'].ffill()

df
```

```python
# 8. Из вышеприведённой таблицы видно, что местоположение даты имеет закономерность. Она выше названия услуги на 1 строку.
# Заполним колонку 'date' с соответствующим смещением.
df['date'] = df.diagnosis.shift(1)
df
```

```python
# 9. Удалим все строки, имеющие хотя бы одно NAN, и обновим индекс.
df = df.dropna().reset_index(drop=True)
df
```

```python
# 10. Приведём 3 последних столбца к требуемому виду.
columns = ['price', 'discounted_sum', 'quantity']
df[columns] = df[columns].astype('int64')
```

```python
# 11. Финальное переименование
df.columns = target_df.columns  
```

```python
# 14. Сравним полученный результат с ожидаемым.
# Результат
df
```

```python
# Загружаем данные
url = 'https://docs.google.com/spreadsheets/d/1o1OuIdZWt-1mrEPCvretwf99w-DdHMPD/edit?usp=drive_link&ouid=104521593033447016990&rtpof=true&sd=true'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]


df_addresses = pd.read_excel(path, sheet_name=1) # адреса, из которых нужно вычленить города, при условии, что они содержаться в перечне городов
df_cities = pd.read_excel(path, sheet_name=0) # перечень городов
```

```python
df_addresses['city'] = None
df_addresses
```

```python
# При импорте df_cities считывается как DataFrame с одним столбцом.
# Поэтому для удобства переведем его в список. 
target_cities = set([с.lower() for с in df_cities['Город']])
target_cities
```

```python
# Вычленим из адресов города
cities = df_addresses['Адреса'].str.lower().apply(lambda x: str(x).split(',')[0].replace('г.', '').replace('г', '').strip())
cities
```

```python
# Наиболее простое решение ❌

i = 0

for c in cities:
    if c in target_cities:
        df_addresses.at[i, 'city'] = c.capitalize()
    i += 1
```

```python
# Результат
df_addresses
```

Цель достигнута. Но, перебор простым циклом не самый оптимальный и эффективный вариант для обработки больших данных. Доработаем! 

Альтернативные варианты:

```python
# Вариант № 1: Избавляемся от счётчика. ❌ 
# enumerate() гарантирует, что i всегда соответствует текущему индексу строки. 

for i, c in enumerate(cities):
    df_addresses.at[i, 'city'] = c.capitalize() if c in target_cities else None

# Но всё ещё есть цикл. Да и метод .at[] сразу вносит изменения в df, что, конечно удобно, но не очень эффективно при большом объёме данных.
```

```python
# Альтернативный вариант № 2. ✅ Более быстрый метод.
df_addresses['city'] = cities.apply(lambda c: c.capitalize() if c in target_cities else None)
```

```python
# Вариант № 3: в его основе лежит генератор списка. Более компактный способ и самый быстрый из всех. ✅✅

df_addresses['city'] = [c.capitalize() if c in target_cities else None for c in cities]
```

```python
# Результат
df_addresses
```
