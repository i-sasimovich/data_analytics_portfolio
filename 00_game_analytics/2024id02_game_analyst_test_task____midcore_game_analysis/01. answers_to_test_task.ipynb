##### –û—Ç–≤–µ—Ç:

```python
""" 
–ß—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–æ –ª–∏ –¥–∞–≤–∞—Ç—å —Ä–µ–∫–ª–∞–º—É, –Ω—É–∂–Ω–æ —Å—Ä–∞–≤–Ω–∏—Ç—å —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (CPI) —Å —Ç–µ–º –¥–æ—Ö–æ–¥–æ–º, –∫–æ—Ç–æ—Ä—ã–π —ç—Ç–æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø—Ä–∏–Ω–µ—Å–µ—Ç –∏–≥—Ä–µ –≤ —Å—Ä–µ–¥–Ω–µ–º, —Ç.–µ. —Å –º–µ—Ç—Ä–∏–∫–æ–π LTV. 
LTV (Life Time Value) - —ç—Ç–æ —Å–æ–≤–æ–∫—É–ø–Ω—ã–π –¥–æ—Ö–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–æ—Å–∏—Ç –æ–¥–∏–Ω –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è, –ø–æ–∫–∞ –æ–Ω –æ—Å—Ç–∞–µ—Ç—Å—è –∞–∫—Ç–∏–≤–Ω—ã–º. 
–°–º—ã—Å–ª LTV: –ø—Ä–∏–∫–∏–Ω—É—Ç—å —Å–∫–æ–ª—å–∫–æ –º–æ–∂–Ω–æ –ø–æ—Ç—Ä–∞—Ç–∏—Ç—å –Ω–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ, —Ç.–µ. –¥–æ–ª–∂–Ω–æ —Å–æ–±–ª—é–¥–∞—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–µ–µ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ LTV > CPI. 

–§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—á—ë—Ç–∞: 
LTV = ARPU x Lifetime 

Lifetime - —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ—Ç —É—Å—Ç–∞–Ω–æ–≤–∫–∏ –¥–æ –Ω–µ–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏. –§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—á—ë—Ç–∞: Lifetime = 1 / Churn  ‚û°Ô∏è LTV = ARPU √ó (1 √∑ Churn) 

Churn - –º–µ—Ç—Ä–∏–∫–∞ –æ—Ç—Ç–æ–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä–∞—è –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç –¥–æ–ª—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–µ–∫—Ä–∞—Ç–∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ç–∞ –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π –ø–µ—Ä–∏–æ–¥ –≤—Ä–µ–º–µ–Ω–∏. 
Churn - –æ–±—Ä–∞—Ç–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ Retention, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ Churn = 1 - Retention.

"""

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
# ARPU (Average Revenue Per User) - —Å—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥ —Å –æ–¥–Ω–æ–≥–æ –∞–∫—Ç–∏–≤–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∑–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π –ø–µ—Ä–∏–æ–¥ –≤—Ä–µ–º–µ–Ω–∏ (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ –∑–∞ –º–µ—Å—è—Ü).(RUB). 
arpu = 46

# 30-day retention - –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –≤–µ—Ä–Ω—É–ª–∏—Å—å –Ω–∞ 30 –¥–µ–Ω—å –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–≥–æ –≤—Ö–æ–¥–∞ –≤ –∏–≥—Ä—É. 
retention_d30 = 0.25

# CPI (Cost Per Install) - –º–µ—Ç—Ä–∏–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–π —É—Å—Ç–∞–Ω–æ–≤–∫–∏ –∏–≥—Ä—ã.(RUB).
cpi = 61

# –†–∞—Å—á—ë—Ç:

churn = 1 - retention_d30

ltv = arpu / churn

if ltv > cpi:
    print(f"LTV —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(ltv,2)} RUB, —á—Ç–æ –ø—Ä–µ–≤—ã—à–∞–µ—Ç CPI –Ω–∞ {round(ltv - cpi, 2)} RUB. –í–æ–∑–º–æ–∂–Ω–æ –¥–∞–≤–∞—Ç—å —Ä–µ–∫–ª–∞–º—É –∏ –∏–º–µ–µ—Ç —Å–º—ã—Å–ª.")
else:
    print(f"LTV —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(ltv,2)}. –î–∞–≤–∞—Ç—å —Ä–µ–∫–ª–∞–º—É –Ω–µ –∏–º–µ–µ—Ç —Å–º—ã—Å–ª–∞.")
```

```python
# C. –û–¥–Ω–æ–¥–Ω–µ–≤–Ω—ã–π –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π —Ä–µ—Ç–µ–Ω—à–Ω (–∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–µ —É–¥–µ—Ä–∂–∞–Ω–∏–µ).

# 1. CTE user_activity:
# - –°–æ–µ–¥–∏–Ω—è–µ–º —Ç–∞–±–ª–∏—Ü—ã users –∏ dailyusers –ø–æ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (uid).
# - –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –º–µ–∂–¥—É –¥–∞—Ç–æ–π —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ (reg_date) –∏ –¥–∞—Ç–æ–π –≤–∏–∑–∏—Ç–∞ (visit_date).
# - –ü–æ—Å–∫–æ–ª—å–∫—É –≤ —Ç–∞–±–ª–∏—Ü–µ `dailyusers` –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω–∞ —É–Ω–∏–∫–∞–ª—å–Ω–æ—Å—Ç—å –∑–∞–ø–∏—Å–µ–π –ø–æ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ uid –∏ day, –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –∞–≥—Ä–µ–≥–∞—Ü–∏—è –¥–ª—è —É—Å—Ç—Ä–∞–Ω–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è.
# - –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø–æ–¥–∑–∞–ø—Ä–æ—Å–∞ –ø–æ–ª—É—á–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Å user_id, reg_date, visit_date, –∏ days_since_reg (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –æ—Ç —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏).
# 2. –û—Å–Ω–æ–≤–Ω–æ–π –∑–∞–ø—Ä–æ—Å. –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –æ–¥–Ω–æ–¥–Ω–µ–≤–Ω–æ–µ —É–¥–µ—Ä–∂–∞–Ω–∏–µ (RR1) –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –ø–æ –¥–∞—Ç–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏.:
# - –ò—Å–ø–æ–ª—å–∑—É–µ–º —É—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä CASE –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –≤–µ—Ä–Ω—É–ª–∏—Å—å –Ω–∞ 1-–π –¥–µ–Ω—å –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏.
# - –î–µ–ª–∏–º —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞ –æ–±—â–µ–µ —á–∏—Å–ª–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ä–∞–º–∫–∞—Ö –∫–æ–≥–æ—Ä—Ç—ã –∏ —É–º–Ω–æ–∂–∞–µ–º –Ω–∞ 100, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –ø—Ä–æ—Ü–µ–Ω—Ç.


# 1–∞—è –≤–µ—Ä—Å–∏—è –∑–∞–ø—Ä–æ—Å–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º CTE ‚ùå
query_c1 = """
WITH user_activity AS (
     SELECT u.uid AS user_id, 
            u.reg_time::date AS reg_date, 
            du.day AS visit_date, 
            (du.day - u.reg_time::date) AS days_since_reg 
      FROM users u
      LEFT JOIN dailyusers du 
        ON u.uid = du.uid
)

SELECT reg_date,
       SUM(CASE WHEN days_since_reg = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(user_id) AS rr1
  FROM user_activity
 GROUP BY reg_date
 ORDER BY reg_date;
"""
# –ò–∑–º–µ–Ω–∏–≤ –∑–Ω–∞—á–µ–Ω–∏–µ WHEN days_since_reg = 1 –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –∑–∞–ø—Ä–æ—Å–µ –Ω–∞ WHEN days_since_reg = 7, –º—ã –ø–æ–ª—É—á–∏–º 7 day Retention.
# –í –∑–∞–ø—Ä–æ—Å–µ –∏–º–µ—é—Ç—Å—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è, —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–º –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –±–æ–ª—å—à–∏–º –æ–±—ä–µ–º–æ–º –¥–∞–Ω–Ω—ã—Ö.
# –ü–æ–ø—Ä–æ–±—É–µ–º –µ–≥–æ –¥–æ—Ä–∞–±–æ—Ç–∞—Ç—å –∏ –æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å, –∏ –ø–æ–ª—É—á–∏–º...


# 2–æ–π –≤–∞—Ä–∏–∞–Ω—Ç –∑–∞–ø—Ä–æ—Å–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º CTE ‚ùå
query_c2 = """
WITH user_activity AS (
     SELECT u.uid AS user_id, 
            u.reg_time::date AS reg_date, 
            du.day AS visit_date
      FROM users u
      LEFT JOIN  dailyusers du 
        ON u.uid = du.uid
)

SELECT reg_date,
       COUNT(CASE WHEN visit_date = reg_date + INTERVAL '1 day' THEN 1 END) * 100.0 / COUNT(user_id) AS rr1
  FROM user_activity
 GROUP BY reg_date
 ORDER BY reg_date;
"""
# –í —ç—Ç–æ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ  —É–±—Ä–∞–Ω—ã –ª–∏—à–Ω–∏–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è, —á—Ç–æ –¥–æ–ª–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –∑–∞–ø—Ä–æ—Å –±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–º. –û–¥–Ω–∞–∫–æ, –≤–æ–∑–º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ CTE –∏–∑–±—ã—Ç–æ—á–Ω–æ.
# –ü–æ–ø—Ä–æ–±—É–µ–º –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç CTE. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø–æ–ª—É—á–∏–º...


# 3–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –∑–∞–ø—Ä–æ—Å–∞ –±–µ–∑ CTE ‚úÖ
query_c3 = """
SELECT u.reg_time::date AS reg_date,
       COUNT(CASE WHEN du.day = u.reg_time::date + INTERVAL '1 day' THEN 1 END) * 100.0 / COUNT(*) AS rr1
  FROM users u 
  LEFT JOIN  dailyusers du 
    ON u.uid = du.uid
 GROUP BY u.reg_time::date
 ORDER BY reg_date;
"""
# 3–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –∑–∞–ø—Ä–æ—Å–∞ - –Ω–∞–∏–ª—É—á—à–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –ø–æ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ –ø—Ä–æ—Å—Ç–æ—Ç–µ, –ø–æ—Å–∫–æ–ª—å–∫—É –æ–Ω –Ω–∞–ø—Ä—è–º—É—é —Ä–µ—à–∞–µ—Ç –∑–∞–¥–∞—á—É –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è CTE –∏ –ª–∏—à–Ω–∏—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π.

# –ö–æ–¥ –ø—Ä–æ–≤–µ—Ä–µ–Ω –∏ —Ä–∞–±–æ—Ç–∞–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –Ω–∞ PostgreSQL 16.4.
```

```python
# D. –û–¥–Ω–æ–¥–Ω–µ–≤–Ω—ã–π —Ä–æ–ª–ª–∏–Ω–≥ —Ä–µ—Ç–µ–Ω—à–Ω (—Å–∫–æ–ª—å–∑—è—â–µ–µ —É–¥–µ—Ä–∂–∞–Ω–∏–µ).

# Rolling Retention –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç –Ω–æ–≤—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤–µ—Ä–Ω—É–ª–∏—Å—å –≤ –ø—Ä–æ–¥—É–∫—Ç –≤ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–π –¥–µ–Ω—å —Å –º–æ–º–µ–Ω—Ç–∞ –ø—Ä–∏—Ö–æ–¥–∞ –∏–ª–∏ –ª—é–±–æ–π –¥–µ–Ω—å –ø–æ—Å–ª–µ –Ω–µ–≥–æ.

# –°–ª–µ–≥–∫–∞ –∏–∑–º–µ–Ω–∏–º –ø—Ä–µ–¥—ã–¥—É—â–∏–π –∑–∞–ø—Ä–æ—Å, –∞ –∏–º–µ–Ω–Ω–æ:
# 1. –ò—Å—Ö–æ–¥—è –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è Rolling Retention, –Ω–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –ø–æ—Å–ª–µ–¥–Ω—é—é –¥–∞—Ç—É –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
# –ü–æ—ç—Ç–æ–º—É –≤ CTE user_activity –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –ø–æ—Å–ª–µ–¥–Ω—é—é/max –¥–∞—Ç—É –≤—Ö–æ–¥–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
# 2. –í –æ—Å–Ω–æ–≤–Ω–æ–º –∑–∞–ø—Ä–æ—Å–µ –∏–∑–º–µ–Ω–∏–º —É—Å–ª–æ–≤–∏–µ –Ω–∞ WHEN days_since_reg >= 1, —á—Ç–æ–±—ã —É—á–µ—Å—Ç—å —Ç–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –≤–µ—Ä–Ω—É–ª–∏—Å—å –Ω–∞ 1-–π –¥–µ–Ω—å –∏–ª–∏ –ø–æ–∑–∂–µ.

# 1–∞—è –≤–µ—Ä—Å–∏—è –∑–∞–ø—Ä–æ—Å–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º CTE ‚ùå
query_d1 = """
WITH user_activity AS (
     SELECT u.uid AS user_id, 
            u.reg_time::date AS reg_date, 
            MAX(du.day) AS last_return_date,  
            (MAX(du.day) - u.reg_time::date) AS days_since_reg
      FROM users u
      LEFT JOIN dailyusers du 
        ON u.uid = du.uid
     GROUP BY u.uid, u.reg_time::date
)

SELECT reg_date,
       SUM(CASE WHEN days_since_reg >= 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(user_id) AS rolling_rr1
  FROM user_activity
 GROUP BY reg_date
 ORDER BY reg_date;
"""
# –í –∑–∞–ø—Ä–æ—Å–µ –∏–º–µ—é—Ç—Å—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è, —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–º –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –±–æ–ª—å—à–∏–º –æ–±—ä–µ–º–æ–º –¥–∞–Ω–Ω—ã—Ö.
# –ü–æ–ø—Ä–æ–±—É–µ–º –µ–≥–æ –¥–æ—Ä–∞–±–æ—Ç–∞—Ç—å –∏ –æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å, –∏ –ø–æ–ª—É—á–∏–º...


# 2–æ–π –≤–∞—Ä–∏–∞–Ω—Ç –∑–∞–ø—Ä–æ—Å–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º CTE ‚úÖ
query_d2 = """
WITH user_activity AS (
     SELECT u.uid AS user_id, 
            u.reg_time::date AS reg_date, 
            MAX(du.day) AS last_return_date  
	   FROM users u
       LEFT JOIN dailyusers du 
         ON u.uid = du.uid
      GROUP BY u.uid, u.reg_time::date
)

SELECT reg_date,
       COUNT(CASE WHEN last_return_date  >= reg_date + INTERVAL '1 day' THEN 1 END) * 100.0 / COUNT(user_id) AS rolling_rr1
  FROM user_activity
 GROUP BY reg_date
 ORDER BY reg_date;
"""
# –í —ç—Ç–æ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ  —É–±—Ä–∞–Ω—ã –ª–∏—à–Ω–∏–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è, —á—Ç–æ –¥–æ–ª–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –∑–∞–ø—Ä–æ—Å –±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–º. –û–¥–Ω–∞–∫–æ, –≤–æ–∑–º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ CTE –∏–∑–±—ã—Ç–æ—á–Ω–æ.
# –ü–æ–ø—Ä–æ–±—É–µ–º –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç CTE. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø–æ–ª—É—á–∏–º...

# 3–∞—è –≤–µ—Ä—Å–∏—è –±–µ–∑ CTE  ‚ùå
query_d3 = """
SELECT u.reg_time::date AS reg_date,
       COUNT(DISTINCT du.uid) * 100.0 / COUNT(DISTINCT u.uid) AS rolling_rr1
  FROM users u
  LEFT JOIN  dailyusers du 
    ON u.uid = du.uid AND du.day > u.reg_time::date 
 GROUP BY u.reg_time::date
 ORDER BY reg_date;

"""
# 3—è—è –≤–µ—Ä—Å–∏—è –∫–∞–∂–µ—Ç—Å—è –±–æ–ª–µ–µ –ø—Ä–æ—Å—Ç–æ–π, –Ω–æ –æ—Ç—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –≤ –¥–≤–æ–µ –¥–æ–ª—å—à–µ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö, –≤–µ—Ä–æ—è—Ç–Ω–æ, –∏–∑-–∑–∞ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –≤ LEFT JOIN –∏ COUNT(DISTINCT ...).
# –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –Ω–∞–∏–±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ–π –≤–µ—Ä—Å–∏–µ–π –±—É–¥–µ—Ç –≤–µ—Ä—Å–∏—è –Ω–æ–º–µ—Ä 2.

# –ö–æ–¥ –ø—Ä–æ–≤–µ—Ä–µ–Ω –∏ —Ä–∞–±–æ—Ç–∞–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –Ω–∞ PostgreSQL 16.4.
```

##### 4. –†–µ—à–∏—Ç–µ –∑–∞–¥–∞—á—É.

–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∫–∞–∫–æ–≤—ã —à–∞–Ω—Å—ã —É –∏–≥—Ä–æ–∫–∞ –ø–æ–ª—É—á–∏—Ç—å 0, 1, 2 –∏–ª–∏ 3 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–∞, –æ—Ç–∫—Ä—ã–≤ –≤—Å–µ 3 —è—â–∏–∫–∞.\

- n  ‚Äî –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π,

```python
# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
p = 0.3
q = 1 - p
n = 3
k0 = 0
k1 = 1
k2 = 2
k3 = 3
```

```python
#1 –†–∞—Å—Å—á–∏—Ç–∞–µ–º  —à–∞–Ω—Å—ã  –∏–≥—Ä–æ–∫–∞ –ø–æ–ª—É—á–∏—Ç—å 0 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤.

# –î–ª—è —ç—Ç–æ–≥–æ —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª–µ –ë–µ—Ä–Ω—É–ª–ª–∏ —Å–Ω–∞—á–∞–ª–∞ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç C(3, 0). 
# –§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—á—ë—Ç–∞: C(n, k) = n! / (k! * (n - k)!)

Cnk0 = (1 * 2 * 3) / (1 * (1 * 2 * 3))
Cnk0

# –≠—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–º–µ–µ—Ç –ø—Ä–æ—Å—Ç–æ–π —Å–º—ã—Å–ª: —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Å–ø–æ—Å–æ–± –≤—ã–±—Ä–∞—Ç—å 0 –æ–±—ä–µ–∫—Ç–æ–≤ –∏–∑ 3 –¥–æ—Å—Ç—É–ø–Ω—ã—Ö, –∏ —ç—Ç–æ "–≤—ã–±—Ä–∞—Ç—å –Ω–∏—á–µ–≥–æ".

# –º–æ–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å–µ–±—è
# import math 
# Cnk0 = math.comb(n, k0) 
```

```python
# –¢–µ–ø–µ—Ä—å —É –Ω–∞—Å –∏–º–µ—é—Ç—Å—è –≤—Å–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ –¥–∞–Ω–Ω—ã–µ. –ü–æ–¥—Å—Ç–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏—è.
probability_0 = Cnk0 * (p ** k0) * ((1 - p) ** (n - k0))
probability_0

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å {k0} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(probability_0,3)}.") 
```

```python
#2 –¢–µ–ø–µ—Ä—å –ø–æ—Å—Ç—É–ø–∏–º –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏ –∏ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º  —à–∞–Ω—Å—ã  –∏–≥—Ä–æ–∫–∞ –ø–æ–ª—É—á–∏—Ç—å 1 —É–Ω–∏–∫–∞–ª—å–Ω—ã–π –ø—Ä–µ–¥–º–µ—Ç.

# –°–Ω–æ–≤–∞ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç C(3, 1)
Cnk1 = (1 * 2 * 3) / (1 * (1 * 2))
Cnk1
```

```python
probability_1 = Cnk1 * (p ** k1) * ((1 - p) ** (n - k1))
probability_1

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å {k1} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(probability_1,3)}.") 
```

```python
#3. –ü–æ –∞–Ω–∞–ª–æ–≥–∏–∏  —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º  —à–∞–Ω—Å—ã  –∏–≥—Ä–æ–∫–∞ –ø–æ–ª—É—á–∏—Ç—å 2 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–∞.
# –°–Ω–æ–≤–∞ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç C(3, 2)
Cnk2 = (1 * 2 * 3) / ((1*2) * (1))
Cnk2
```

```python
probability_2 = Cnk2 * (p ** k2) * ((1 - p) ** (n - k2))
probability_2

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å {k2} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(probability_2,3)}.") 
```

```python
#3.–ò –Ω–∞–∫–æ–Ω–µ—Ü  —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º  —à–∞–Ω—Å—ã  –∏–≥—Ä–æ–∫–∞ –ø–æ–ª—É—á–∏—Ç—å 3 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–∞.
# –°–Ω–æ–≤–∞ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç C(3, 3)
Cnk3 = (1 * 2 * 3) / ((1*2*3) * (1))
Cnk3
```

```python
probability_3 = Cnk3 * (p ** k3) * ((1 - p) ** (n - k3))
probability_3

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å {k3} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {round(probability_3,3)}.") 
```
