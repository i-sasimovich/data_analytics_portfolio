–ù–µ —É—Å–ø–µ–ª —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ, –∑–∞–ª–∏–ª –ø–æ 20 —Å—Ç—Ä–æ–∫ ü•≤

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–∞–ø—Ä–æ—Å–æ–≤
def select(sql: str) -> pd.DataFrame:
    result = client.query(sql)
    return pd.DataFrame(result.result_rows, columns=result.column_names)
```

1) –∫–æ—Ç–æ—Ä—ã–π –ø–æ—Å—á–∏—Ç–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —É—Å—Ç—Ä–æ–π—Å—Ç–≤ –∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –º–µ—Å—è—Ü–∞–º –≤ –ø–µ—Ä–∏–æ–¥ —Å –ò—é–ª—è-2023 –ø–æ –î–µ–∫–∞–±—Ä—å-2023 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ);

```python
query_1 = """
SELECT
  toYYYYMM(event_timestamp) AS month,
  COUNT(DISTINCT gadget_id) AS unique_devices,
  COUNT(DISTINCT player_id) AS unique_users
FROM event_log
WHERE event_timestamp BETWEEN '2023-07-01' AND '2023-12-31'
GROUP BY month
ORDER BY month;
""".strip()

select(query_1)
```

2) –∫–æ—Ç–æ—Ä—ã–π –≤—ã–≤–µ–¥–µ—Ç –¢–û–ü-5 —Å—Ç—Ä–∞–Ω –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ 2023 –≥–æ–¥—É;

```python
query_2 = """
SELECT
  country_code,
  COUNT(player_id) AS user_count
FROM user_info
WHERE toYear(record_created) = '2023'
GROUP BY country_code
ORDER BY user_count DESC
LIMIT 5;
""".strip()

select(query_2)
```

3) –∫–æ—Ç–æ—Ä—ã–π –≤—ã–≤–µ–¥–µ—Ç –¢–û–ü-5 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –æ–±—â–µ–π —Å—É–º–º–µ –ø–ª–∞—Ç–µ–∂–µ–π, –∏ –ø–æ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–∞–Ω–µ –∑–∞ 2023 –≥–æ–¥;

```python
query_3 = """
WITH ranked_users AS (
  SELECT
    country_code,
    player_id,
    SUM(amount_RUB) AS total_payment,
    ROW_NUMBER() OVER (PARTITION BY country_code ORDER BY SUM(amount_RUB) DESC) AS rank
  FROM payment_records
  JOIN user_info USING (player_id)
  WHERE payment_successful = 1 AND toYear(payment_completed) = '2023'
  GROUP BY country_code, player_id
)
SELECT country_code, player_id, total_payment
FROM ranked_users
WHERE rank <= 5;
""".strip()

select(query_3)
```

4) –∫–æ—Ç–æ—Ä—ã–π –ø–æ—Å—á–∏—Ç–∞–µ—Ç —Å—Ä–µ–¥–Ω—é—é –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–µ—Å—Å–∏–∏ –Ω–∞ —Å–∞–π—Ç–µ –∑–∞ 2023 –≥–æ–¥;

```python
query_4 = """
SELECT 
   AVG(session_duration) AS avg_session_duration
FROM (
  SELECT 
    play_session_id, 
    MAX(event_timestamp) - MIN(event_timestamp) AS session_duration
  FROM event_log
  WHERE toYear(event_timestamp) = 2023
  GROUP BY play_session_id
) AS sessions;
""".strip()

select(query_4)
```

5) –∫–æ—Ç–æ—Ä—ã–π –ø–æ—Å—á–∏—Ç–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø–æ—Å–µ—Ç–∏–≤—à–∏—Ö —Å–∞–π—Ç (DAU) –ø–æ –º–µ—Å—è—Ü–∞–º, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–ª–∏ —Å–≤–æ–π email-–∞–¥—Ä–µ—Å. –ê–Ω–∞–ª–∏—Ç–∏–∫–∞ –Ω—É–∂–Ω–∞ –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å –ò—é–ª—è-2023 –ø–æ –î–µ–∫–∞–±—Ä—å-2023 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ);

```python
query_5 = """
SELECT
  toYYYYMM(event_timestamp) AS month,
  COUNT(DISTINCT player_id) AS MAU
FROM event_log
JOIN user_info USING (player_id)
WHERE event_timestamp BETWEEN '2023-07-01' AND '2023-12-31'
  AND email_verified != '1970-01-01'
GROUP BY month
ORDER BY month;
""".strip()

select(query_5)

# –ö–∞–∫ —è –ø–æ–Ω—è–ª, –≤ –∑–∞–¥–∞–Ω–∏–∏ –∏–º–µ–ª–æ—Å—å –≤–≤–∏–¥—É –Ω–µ DAU, –∞ MAU.
```

6) –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª–∏—Ç –æ—Ü–µ–Ω–∏—Ç—å –¥–æ–ª—é (–ø–æ —Å—É–º–º–µ —É—Å–ø–µ—à–Ω–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞) –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–Ω–æ–π –ø–æ—á—Ç–æ–π, –≤ –î–µ–∫–∞–±—Ä–µ 2023 –≥–æ–¥–∞;

```python
query_6 = """
SELECT
  SUM(CASE WHEN email_verified != '1970-01-01' THEN amount_RUB ELSE 0 END) / SUM(amount_RUB) AS verified_user_share
FROM payment_records
JOIN user_info USING (player_id)
WHERE payment_successful = 1
  AND payment_completed >= '2023-12-01' 
  AND payment_completed < '2024-01-01'
""".strip()

select(query_6)
```

7) –∫–æ—Ç–æ—Ä—ã–π –≤—ã–≤–µ–¥–µ—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –Ω–∏–∂–Ω–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å, –º–µ–¥–∏–∞–Ω—É –∏ –≤–µ—Ä—Ö–Ω–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –º–µ–∂–¥—É —É—Å–ø–µ—à–Ω—ã–º–∏ –¥–µ–ø–æ–∑–∏—Ç–∞–º–∏.

```python
query_7 = """
SELECT
  player_id,
  AVG(time_between_deposits) / 3600 AS avg_time_between_deposits_hours,
  quantile(0.25)(time_between_deposits) / 3600 AS lower_quartile_hours,
  quantile(0.5)(time_between_deposits) / 3600 AS median_time_hours,
  quantile(0.75)(time_between_deposits) / 3600 AS upper_quartile_hours
FROM (
  SELECT
    player_id,
    dateDiff('second', 
             lagInFrame(payment_completed) OVER (PARTITION BY player_id ORDER BY payment_completed),
             payment_completed) AS time_between_deposits
  FROM payment_records
  WHERE payment_successful = 1
    AND payment_completed != '1970-01-01 00:00:00'
) AS time_data
WHERE time_between_deposits > 0
GROUP BY player_id
HAVING COUNT(time_between_deposits) > 1; 
""".strip()

select(query_7)

# HAVING COUNT(time_between_deposits) > 1 --- –¥.–±. –º–∏–Ω–∏–º—É–º 2 —É—Å–ø–µ—à–Ω—ã—Ö –¥–µ–ø–æ–∑–∏—Ç–∞
```

```python
df = pd.read_parquet(r'D:\000_Jupyter_notebooks\2025 02 Maxbitsolution, Data-analyst\task_3_events.parquet')
df.head()
```

```python
df = df.sort_values(by=['user_id', 'time'])
```

```python
# –°–¥–≤–∏–≥–∞–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —Ä–∞–∑—Ä—ã–≤–æ–≤ –º–µ–∂–¥—É –¥–µ–π—Å—Ç–≤–∏—è–º–∏
df['prev_time'] = df.groupby('user_id')['time'].shift()
```

```python
# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ä–∞–∑—Ä—ã–≤—ã –º–µ–∂–¥—É –¥–µ–π—Å—Ç–≤–∏—è–º–∏ (–≤ –º–∏–Ω—É—Ç–∞—Ö)
df['time_diff'] = (df['time'] - df['prev_time']).dt.total_seconds() / 60
```

```python
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –Ω–∞—á–∞–ª–æ –Ω–æ–≤–æ–π —Å–µ—Å—Å–∏–∏ (—Ä–∞–∑—Ä—ã–≤ > 30 –º–∏–Ω—É—Ç –∏–ª–∏ –ø–µ—Ä–≤–æ–µ —Å–æ–±—ã—Ç–∏–µ)
df['new_session'] = (df['time_diff'] > 30) | (df['prev_time'].isna())
```

```python
# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –º–µ—Ç–∫—É –¥–ª—è –∫–∞–∂–¥–æ–π —Å–µ—Å—Å–∏–∏
df['session_id'] = df.groupby('user_id')['new_session'].cumsum()
```

```python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–ª—è —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è —Å–µ—Å—Å–∏–π
session_df = df.groupby(['user_id', 'user_group', 'session_id']).agg(
    session_start=('time', 'min'),
    session_end=('time', 'max')
).reset_index()
```

```python
session_df['session_date'] = session_df['session_start'].dt.date
```

```python
session_df = session_df.rename(columns={'user_group': 'ab_group'})
```

```python
session_df = session_df.drop(columns=['session_id'])
```

```python
# –ü—Ä–æ–≤–µ—Ä–∏–º –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –¥—É–±–ª–∏—Ä—É—é—â–∏—Ö —Å—Ç—Ä–æ–∫ 
session_df.duplicated().sum()
```

```python
# –ü—Ä–æ–≤–µ—Ä–∏–º –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
session_df.isna().sum()
```

```python
session_counts = session_df.groupby(['user_id', 'ab_group'])['session_date'].nunique().reset_index()
session_counts.columns = ['user_id', 'ab_group', 'session_count']
session_counts.head()
```

```python
session_counts['converted'] = session_counts['session_count'] > 1
session_counts.head()
```

```python
conversion_rates = session_counts.groupby('ab_group')['converted'].mean().reset_index()
conversion_rates.columns = ['ab_group', 'conversion_rate']
conversion_rates

# –ö–æ–Ω—Ç—Ä–æ–ª—å–Ω–∞—è –≥—Ä—É–ø–ø–∞ –ê: ab_group = 0
# –¢–µ—Å—Ç–æ–≤–∞—è –≥—Ä—É–ø–ø–∞ B: ab_group = 1
```

–†–µ–∑—É–ª—å—Ç–∞—Ç: –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤–æ –≤—Ç–æ—Ä—É—é —Å–µ—Å—Å–∏—é —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 19.44% –¥–ª—è –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø—ã –∏ 19.00% –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π. –°–ª—É—á–∞–π–Ω–∞ –ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞, –Ω–µ –æ—á–µ–≤–∏–¥–Ω–æ. –¢—Ä–µ–±—É–µ—Ç—Å—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–Ω–∏—Ü—ã.

–ü—Ä–∏–º–µ–Ω–µ–Ω–∏–º —Ç–µ—Å—Ç —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç.

–û–ø—Ä–µ–¥–µ–ª–∏–º—Å—è —Å –≥–∏–ø–ø–æ—Ç–µ–∑–∞–º–∏:

* H0: –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –Ω–µ –∑–Ω–∞—á–∏–º–∞. –í—ã–±–æ—Ä–∫–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –æ–¥–Ω–æ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, —Ä–∞–∑–ª–∏—á–∏–π –Ω–µ—Ç, –Ω–æ–≤–æ–≤–≤–µ–¥–µ–Ω–∏—è –Ω–µ –æ–∫–∞–∑—ã–≤–∞—é—Ç –≤–ª–∏—è–Ω–∏—è.

* H1: –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞. –†–∞–∑–ª–∏—á–∏—è –µ—Å—Ç—å.

```python
from scipy.stats import chi2_contingency
```

```python
conversion_data = session_counts.groupby('ab_group')['converted'].value_counts().unstack().fillna(0)
conversion_data 
```

```python
# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–∞ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2_stat, p_value, dof, expected = chi2_contingency(conversion_data)

print(f"\n–•–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {chi2_stat:.4f}")
print(f"P-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value:.4f}")

if p_value < 0.05:
    print("–†–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞.")
else:
    print("–†–∞–∑–Ω–∏—Ü–∞ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–π.")
```

P-–∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Ç–µ—Å—Ç–∞ —Å–æ—Å—Ç–∞–≤–∏–ª–æ 0.5151, —á—Ç–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–µ–≤—ã—à–∞–µ—Ç —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –Ω–∞–±–ª—é–¥–∞–µ–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ —Ç–µ—Å—Ç–æ–≤–æ–π (19.00%) –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π (19.44%) –≥—Ä—É–ø–ø –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–±—ä—è—Å–Ω–µ–Ω–∞ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –∫–æ–ª–µ–±–∞–Ω–∏—è–º–∏, –∞ –Ω–µ —Ä–µ–∞–ª—å–Ω—ã–º —ç—Ñ—Ñ–µ–∫—Ç–æ–º –Ω–æ–≤–æ–≤–≤–µ–¥–µ–Ω–∏–π.

–ü—Ä–∏ —Ç–∞–∫–æ–º –∏—Å—Ö–æ–¥–µ –±–∏–∑–Ω–µ—Å—É —Å—Ç–æ–∏—Ç —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ —à–∞–≥–∏:

* –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –¥—Ä—É–≥–∏–µ –≥–∏–ø–æ—Ç–µ–∑—ã, –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–µ –Ω–∞ —É–ª—É—á—à–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –ø–æ–≤–µ–¥–µ–Ω–∏—è.
* –ü—Ä–æ–≤–µ—Å—Ç–∏ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Å –±–æ–ª–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–º–∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏, —Ç–∞–∫–∏–º–∏ –∫–∞–∫ –≤—Ä–µ–º—è –º–µ–∂–¥—É —Å–µ—Å—Å–∏—è–º–∏, –≤–æ–≤–ª–µ—á–µ–Ω–Ω–æ—Å—Ç—å –∏–ª–∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ (ARPU).
